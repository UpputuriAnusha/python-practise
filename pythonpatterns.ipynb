```python
#Right angle triangled pattern
rows=int(input("enter the row size for pattern:"))
for i in range(1,rows+1):
  for j in range(1,i+1):
    print("*",end=" ")
  print()
```

```python
#Inverted right angled triangle pattern
rows=int(input("enter the row size for pattern:"))
for i in range(rows,0,-1):
  for j in range(1,i+1):
    print("*",end=" ")
  print()
```

```python
#Pyramid pattern
rows=int(input("enter the row size for pattern:"))
for i in range(1,rows+1):
  for j in range(rows-i):
    print(" ",end=" ")
  for k in range(1,2*i):
    print("*",end=" ")
  print()
```

```python
rows = int(input("Enter the size for pattern: "))
# Upper part
for i in range(1, rows + 1):
    for j in range(rows - i):
        print(" ", end=" ")
    for k in range(1, 2 * i):
        if k == 1 or k == 2 * i - 1:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()

# Lower part
for i in range(rows - 1, 0, -1):
    for j in range(rows - i):
        print(" ", end=" ")
    for k in range(1, 2 * i):
        if k == 1 or k == 2 * i - 1:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
```
