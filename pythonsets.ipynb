```python
#crate and print set
s={5,10,15,20}
print(s)
```

```python
#add element to set
s={5,10,15,20}
s.add(20)
print(s)
```

```python
#updating the set with multiple elements
s={5,10,15,20}
s.update([25,36])
print(s)
```

```python
#remove an element in the set
s={5,10,15,20}
s.discard(10)
s.discard(22) # Using discard() instead of remove()
print(s)
```

```python
#discarding the set
s={5,10,15,20}
s.discard(22)
s.discard(5)
print(s)
```

```python
#set union
s={5,10,15,20}
u={2,4,6,8}
print(s|u)
```

```python
#set intersection
s={5,10,15,20}
u={2,4,5,20}
print(s&u)
```

```python
#check is subset or superset
a={10}
b={10,20,30}
print(a.issubset(b))
print(b.issuperset(a))
```

```python
#removing all elements from a set
s={5,10,15,20}
s.clear()
print(s)
```
