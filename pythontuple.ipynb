```python
t=(10,20,30,40,50)
print(t)
```

```python
#accessing tple elements
t=(10,20,30,40,50)
print(t[4])
```

```python
# length of tuple
t=(10,20,30,40,50)
print(len(t))
```

```python
# finding max and min value of tuple
t=(10,20,30,40,50)
print(max(t))
print(min(t))
```

```python
#finding index of anelement in tuple
t=(10,20,30,40,50)
print(t.index(40))
```

```python
#checking if an element exist in tuple or not
t=(10,20,30,40,50)
print( 30 in t)
```

```python
#concatenating of two tuples
t1=(10,20,30,40,50)
t2=(100,200,300,400,500)
t3=t1+t2
print(t3)
```

```python
#coverting list into tuple
a=[20,30,40,50,60]
b=tuple(a)
print(b)
```

```python
#count occurance of an element in tuple
t=(10,20,30,30,50,40,50)
print(t.count(30))
```

```python
# converting tuple to dictionary
t=(("a",12),("b",39))
d=dict(t)
print(d)
```

```python
#comparing  two tuple
t1=(20,30,40,50)
t2=(20,30,40,50)
print(t1==t2)
```

```python
#sorting two tuples
t=(1,5,9,33,30,88)
sorted_tuple=tuple(sorted(t))
print(sorted_tuple)
```
