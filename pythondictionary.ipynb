```python
#create and printing a dictionary
d={"name":"anusha","age":20}
print(d)
print(type(d))
```

```python
#updating a dictionary
d={"name":"anusha","age":20}
d["address"]="darsi"
print(d)
```

```python
#adding a new key_value  pair
d={"name":"anusha","age":20}
d["email"]="anushaupputuri34@gmail.com"
print(d)
```

```python
#deleting a key value pair
d={"name":"anusha","age":20}
del d["age"]
print(d)
```

```python
#check if key exists or not
d={"name":"anusha","age":20}
print('age' in d)
```

```python
#getting all keys and values
d={"name":"anusha","age":20,"address":"rajampalli"}
print(d.keys())
print(d.values())
```

```python
#copying a dictionary
d={"name":"anusha","age":20,"address":"rajampalli"}
d1=d.copy()
print(d1)
```

```python
#finding a length of dictionary
d={"name":"anusha","age":20,"address":"rajampalli"}
print(len(d))
```

```python
#merging two dictinarys
d1={"a":1}
d2={"b":2}
d1.update(d2)
print(d1)
```
